(1/2) ***Your answer goes here:***

The main assumption behind Naive Bayes is the conditional independence assumption, which states that the features are mutually independent, conditional on the class labels. This allows us to factor the class conditional likelyhood $P(x_1, x_2, \dots, x_n | C)$ as $P(x_1 | C) P(x_2 | C) \dots P(x_n | C)$.

This way, each $P(x_i | C)$ can be estimated independently with a reasonable amount of data. Without this assumption the number of training examples needed to estimate $P(x_1, x_2, \dots, x_n | C)$ well is exponential in $n$. Therefore, for all but very small values of $n$ we would require more data to train our our model than we could ever hope to collect.

However, I suspect that one of the main reasons for the dramatic improvement is in fact data snooping. Here we analysed the full dataset and decided to clip the outliers based on that. We then subsequently ran K-Fold cross-validation. This is not good practice because the outliers which were clipped will also appear in the validation sets of some folds. When performing K-Fold cross-validation any pre-processing steps which involve looking at the data should be done inside the cross-validation loop. 

So in this case, the correct way to do it would be to analyse only the training set for each of the K splits and remove outliers from the training set only. We should not analyse or modify any of the K validation sets in any way.
